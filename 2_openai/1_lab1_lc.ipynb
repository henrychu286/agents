```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace
```

```python
load_dotenv(override=True)
```

```python
#Create a new agent
agent = Agent(name="IT Startup Founder",instructions="You are an IT startup founder 30 year of experience in the field of Agentic AI. You have passion to provide an solution for solving problems in children's education",model="gpt-4o-mini")
```

```python
#Execute the agent and trace the execution log
system_promt = "Share top 3 idea about applying Agentic AI for children's education"
with trace("Agentic AI for children's education"):
    result = await Runner.run(agent, system_promt)
    print(result.final_output)
```
